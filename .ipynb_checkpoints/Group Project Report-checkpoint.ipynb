**Introduction**

- ***Background Information***

  The project focus on the students' knowledge level of Electrical DC Machine. Electrical DC Machine is an electromechanical device that is used to convert electrical energy into mechanical energy or vice versa. There are 4 levels of knowledge status created by the author of the database. They collected five input values in order to get one out put value of interest, and two of those input values involves exam performance. It is clear that is requires lot of work to collect all these information. Therefore our group wants to try predict the UNS of the observations with less input values which are easier to collect. After creating visualization of the relationship between the input value and the output value, we observed that the input value PEG (The exam performance of user for goal objects) is crucial for us in order to determine the value of determine the output value UNS (The knowledge level of user). 

   We are trying to predict the UNS (The knowledge level of user) of new observations with STG (The degree of study time for goal object materials) and PEG (The exam performance of user for goal objects).


- ***Question to solve***

  Classify the user's knowledge level of Electric DC Machine with STG and PEG.


- ***Database description***

  Our group is going to use the dataset User Knowledge Modeling. The authors of the dataset conducted surveys to collect relevant data and classified the students’ knowledge status about the subject of Electrical DC Machines. To determine the UNS (The knowledge level of user) of the student, the authors recorded the information of the students in the area of STG (The degree of study time for goal object materials), SCG (The degree of repetition number of user for goal object materials), STR (The degree of study time of user for related objects with goal object), LPR (The exam performance of user for related objects with goal object) and PEG (The exam performance of user for goal objects). The users' knowledge class were classified by the authors using intuitive knowledge classifier (a hybrid ML technique of k-NN and meta-heuristic exploring methods), k-nearest neighbor algorithm.


Import the library we are going to use.

```r
library(tidyverse)
library(repr)
library(tidymodels)
library(readxl)
```

Read the dataset that we are going to explore from the internet int the format of excel.

```r
dc_data <- read_excel("Data_User_Modeling_Dataset.xlsx", sheet = 2)
```

Modify and mutate the data before analysing it.

```r
tidy_data <- dc_data |>
    select(STG, PEG, UNS) |>
    mutate(UNS = as_factor(UNS))


head(tidy_data)
```

Now we will create a plot to see how the two variables relate to eachother.


```r
dc_plot <- tidy_data |>
    ggplot(aes(x=STG, y=PEG, color=UNS))+
    geom_point()+
    labs(x="The degree of study time for goal object materials", 
         y="The exam performance of users for goal objects", color="Knowledge Level") +
    ggtitle("The DC knowledge level Plot")

dc_plot
```

We will now set seed and split the data into a training and testing set. The training data will take up 75% of the data that we have.

```r
set.seed(2000)

dc_split <- initial_split(tidy_data, prop = 0.75, strata = UNS)  
dc_train <- training(dc_split)   
dc_test <- testing(dc_split)
```

We will now create a recipe and scale the data. 

```r
dc_recipe <- recipe(UNS ~ STG + PEG , data = dc_train) |>
    step_scale(all_predictors()) |>
    step_center(all_predictors()) 
```

Performing the 5-fold cross-validation on the training data.

```r
dc_vfold <- vfold_cv(dc_train, v = 5, strata = UNS)

dc_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |>
       set_engine("kknn") |>
       set_mode("classification")


dc_results<- workflow() |>
       add_recipe(dc_recipe) |>
       add_model(dc_spec) |>
       tune_grid(resamples = dc_vfold, grid = 10) |>
       collect_metrics()

dc_results
```

We plot the accuracy of each value of K to find out the best value of number of neighbors.

```r
accuracies <- dc_results |> 
       filter(.metric == "accuracy")

accuracy_versus_k <- ggplot(accuracies, aes(x = neighbors, y = mean))+
       geom_point() +
       geom_line() +
       labs(x = "Neighbors", y = "Accuracy Estimate") +
       scale_x_continuous(breaks = seq(0, 14, by = 1)) +  # adjusting the x-axis
       scale_y_continuous(limits = c(0.4, 1.0)) # adjusting the y-axis

accuracy_versus_k
```

 We can see that the best values of K are 3 and 4 with the same accuracy value. We tried both K=3 and K=4. It turns out they produce the exact same value. We will take 3 as our K to classify from now on because the program would be more effecient with less values to compute.

```r
dc_final_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = 3) |>
       set_engine("kknn") |>
       set_mode("classification")

dc_final_fit <- workflow() |>
       add_recipe(dc_recipe) |>
       add_model(dc_final_spec) |>
       fit(data = dc_train)
```

```r
dc_test_predictions <- predict(dc_final_fit , dc_test) |>
       bind_cols(dc_test)

dc_test_predictions
```

```r
dc_prediction_accuracy <- dc_test_predictions |>
         metrics(truth = UNS, estimate = .pred_class)

dc_prediction_accuracy
```

```r
dc_mat <- dc_test_predictions |> 
       conf_mat(truth = UNS, estimate = .pred_class)


dc_mat
```

**Discussion**

- **Summary**

  By using only STG and PEG, we can classify the UNS of the observations with the maximum accuracy of 78.8%. The number of the nearest neighbors which lead to the most accurate result is 3.
  
  The class of the UNS is greatly influenced by the value of PEG. All the other values only have minor influence on the UNS. 


- **Discussion of the expectation**

  Before the project, we were expecting to find the UNS in a simpler way which uses less variables so that the data is easier to collect. The attempting of finding new UNS from only STG and PEG is meaningful since it classified most of the new observations correctly. 


- **Impact of the findings**
  
  Based on what we already have in our database, we can predict the UNS of new observations by perform classification with the values of STG and PEG with the accuracy of approximately 78.8%.
